```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('df_clean_exposure.csv').drop(['Unnamed: 0'], axis=1)
```

```python
data_bootstrapped = data.copy()
```

```python
data_participant_means = data.groupby(['participant', 'block', 'node type','condition']).mean(numeric_only=True).reset_index()
```

```python
sns.catplot(
    x = 'block', 
    y = 'data_bootstrapped', 
    hue = 'condition',
    col = 'node type',
    data = data_participant_means,
    kind = 'point',
    hue_order = ['unstructured', 'structured'],
    # errorbar =  ('pi', 95)
)
```
